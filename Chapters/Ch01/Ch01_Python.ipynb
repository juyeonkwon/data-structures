1.5 Ïó¨Îü¨ Í∞úÏùò Í∞íÏùÑ Î∞òÌôò

* ÌÇ§Îäî Í≥†Ïú†Ìï¥Ïïº ÌïòÎ©∞, Î≥ÄÍ≤Ω Î∂àÍ∞ÄÎä•Ìïú Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖÏù¥Ïñ¥Ïïº Ìï©ÎãàÎã§ (Î≥¥ÌÜµ Î¨∏ÏûêÏó¥Ïù¥ÎÇò Ïà´ÏûêÍ∞Ä ÎßéÏù¥ ÏÇ¨Ïö©Îê©ÎãàÎã§).

1. Ï†ÑÏó≠ Î≥ÄÏàò (Global Variables)

Ï†ÑÏó≠ Î≥ÄÏàòÎäî Ìï®Ïàò Î∞îÍπ•ÏóêÏÑú Ï†ïÏùòÎêú Î≥ÄÏàòÎ°ú, ÌîÑÎ°úÍ∑∏Îû® Ï†ÑÏ≤¥ÏóêÏÑú Ï†ëÍ∑ºÌï† Ïàò ÏûàÏäµÎãàÎã§.

```python
x = 10  # Ï†ÑÏó≠ Î≥ÄÏàò

def print_x():
    print(x)  # Ìï®Ïàò ÎÇ¥Î∂ÄÏóêÏÑú Ï†ÑÏó≠ Î≥ÄÏàòÏóê Ï†ëÍ∑º

print_x()  # Ï∂úÎ†•: 10

# Ìï®Ïàò Ïô∏Î∂ÄÏóêÏÑú Ï†ïÏùòÎêú Î≥ÄÏàò xÎäî Ï†ÑÏó≠ Î≥ÄÏàòÎ°ú, print_x() Ìï®Ïàò ÎÇ¥Î∂ÄÏóêÏÑúÎèÑ Ï†ëÍ∑ºÌï† Ïàò ÏûàÏäµÎãàÎã§.
```

2. ÏßÄÏó≠ Î≥ÄÏàò (Local Variables)

ÏßÄÏó≠ Î≥ÄÏàòÎäî Ìï®Ïàò ÎÇ¥Î∂ÄÏóêÏÑú Ï†ïÏùòÎêú Î≥ÄÏàòÎ°ú, Ìï¥Îãπ Ìï®Ïàò ÎÇ¥ÏóêÏÑúÎßå Ïú†Ìö®Ìï©ÎãàÎã§.
Ìï®ÏàòÍ∞Ä Ï¢ÖÎ£åÎêòÎ©¥ ÏßÄÏó≠ Î≥ÄÏàòÎäî ÏÜåÎ©∏ÌïòÎ©∞, Ìï®Ïàò Ïô∏Î∂ÄÏóêÏÑúÎäî Ï†ëÍ∑ºÌï† Ïàò ÏóÜÏäµÎãàÎã§

```python
y = 10
def my_function():
    y = 5  # ÏßÄÏó≠ Î≥ÄÏàò
    print(y)

my_function()  # Ï∂úÎ†•: 5
print(y)  # Ïò§Î•ò Î∞úÏÉù: NameError: name 'y' is not defined

# my_function() ÎÇ¥Î∂ÄÏóêÏÑú Ï†ïÏùòÎêú Î≥ÄÏàò yÎäî ÏßÄÏó≠ Î≥ÄÏàòÏù¥Î©∞, Ìï®Ïàò Ïô∏Î∂ÄÏóêÏÑúÎäî ÏÇ¨Ïö©Ìï† Ïàò ÏóÜÍ∏∞ ÎïåÎ¨∏Ïóê, Ìï®Ïàò Î∞ñÏóêÏÑú yÎ•º Ï∂úÎ†•ÌïòÎ†§Í≥† ÌïòÎ©¥ Ïò§Î•òÍ∞Ä Î∞úÏÉùÌï©ÎãàÎã§.
```

3. Ï†ÑÏó≠ Î≥ÄÏàòÏôÄ ÏßÄÏó≠ Î≥ÄÏàòÏùò Í¥ÄÍ≥Ñ

Ìï®Ïàò ÎÇ¥ÏóêÏÑú Í∞ôÏùÄ Ïù¥Î¶ÑÏùò Î≥ÄÏàòÍ∞Ä Ï†ïÏùòÎêòÎ©¥, Ìï¥Îãπ Î≥ÄÏàòÎäî ÏßÄÏó≠ Î≥ÄÏàòÎ°ú Ï∑®Í∏âÎê©ÎãàÎã§. Ìï®Ïàò ÎÇ¥ÏóêÏÑúÎßå Ìï¥Îãπ Î≥ÄÏàòÏóê Ï†ëÍ∑ºÌï† Ïàò ÏûàÏúºÎ©∞, Ï†ÑÏó≠ Î≥ÄÏàòÎäî Í∞ÄÎ†§ÏßëÎãàÎã§

```python
x = 10  # Ï†ÑÏó≠ Î≥ÄÏàò

def my_function():
    x = 5  # ÏßÄÏó≠ Î≥ÄÏàò
    print(x)  # ÏßÄÏó≠ Î≥ÄÏàòÍ∞Ä Ï∂úÎ†•Îê®

my_function()  # Ï∂úÎ†•: 5
print(x)  # Ï∂úÎ†•: 10 (Ï†ÑÏó≠ Î≥ÄÏàò)

# Ìï®Ïàò my_function() ÎÇ¥ÏóêÏÑú xÎùºÎäî ÏßÄÏó≠ Î≥ÄÏàòÎ•º Ï†ïÏùòÌñàÍ∏∞ ÎïåÎ¨∏Ïóê, Ìï®Ïàò ÎÇ¥Î∂ÄÏóêÏÑúÎäî ÏßÄÏó≠ Î≥ÄÏàò xÍ∞Ä ÏÇ¨Ïö©ÎêòÍ≥†, Ìï®Ïàò Ïô∏Î∂ÄÏóêÏÑúÎäî Ï†ÑÏó≠ Î≥ÄÏàò xÍ∞Ä ÏÇ¨Ïö©Îê©ÎãàÎã§.
```

4. global ÌÇ§ÏõåÎìú

Ìï®Ïàò ÎÇ¥Î∂ÄÏóêÏÑú Ï†ÑÏó≠ Î≥ÄÏàòÏùò Í∞íÏùÑ Î≥ÄÍ≤ΩÌïòÎ†§Î©¥ global ÌÇ§ÏõåÎìúÎ•º ÏÇ¨Ïö©Ìï¥Ïïº Ìï©ÎãàÎã§.
global ÌÇ§ÏõåÎìúÎäî Ìï®Ïàò ÎÇ¥Î∂ÄÏóêÏÑú Ï†ÑÏó≠ Î≥ÄÏàòÎ•º ÏàòÏ†ïÌï† Ïàò ÏûàÎèÑÎ°ù Ìï©ÎãàÎã§.

```python
x = 10  # Ï†ÑÏó≠ Î≥ÄÏàò

def modify_global():
    global x  # Ï†ÑÏó≠ Î≥ÄÏàò xÎ•º ÏÇ¨Ïö©
    x = 20  # Ï†ÑÏó≠ Î≥ÄÏàòÏùò Í∞íÏùÑ Î≥ÄÍ≤Ω
    print(x)  # Ï∂úÎ†•: 20

modify_global()
print(x)  # Ï∂úÎ†•: 20 (Ï†ÑÏó≠ Î≥ÄÏàòÏùò Í∞íÏù¥ Î≥ÄÍ≤ΩÎê®)
```

5. Î≥ÄÏàòÏùò Î≤îÏúÑ Í∑úÏπô (LEGB Í∑úÏπô)

ÌååÏù¥Ïç¨ÏùÄ Î≥ÄÏàòÎ•º Ï∞æÏùÑ Îïå LEGB Í∑úÏπôÏùÑ Îî∞Î¶ÖÎãàÎã§. Ïù¥Îäî Î≥ÄÏàòÍ∞Ä Ïñ¥Îñ§ Î≤îÏúÑÏóê ÏûàÎäîÏßÄ ÌôïÏù∏Ìï† Îïå Ï∞∏Ï°∞ÌïòÎäî ÏàúÏÑúÎ•º ÏùòÎØ∏Ìï©ÎãàÎã§.

* L (Local): Í∞ÄÏû• ÏïàÏ™Ω Ìï®ÏàòÎÇò ÏΩîÎìú Î∏îÎ°ùÏùò ÏßÄÏó≠ Î≥ÄÏàò.
* E (Enclosing): Ï§ëÏ≤© Ìï®ÏàòÏùò Ïô∏Î∂Ä Ìï®ÏàòÏóê ÏûàÎäî Î≥ÄÏàò.
* G (Global): Î™®ÎìàÏùò Ï†ÑÏó≠ Î≥ÄÏàò.
* B (Built-in): ÌååÏù¥Ïç¨Ïùò ÎÇ¥Ïû• Ìï®ÏàòÎÇò Î≥ÄÏàò.

```python
x = "global"

def outer_function():
    x = "enclosing"

    def inner_function():
        x = "local"
        print(x)  # Í∞ÄÏû• Í∞ÄÍπåÏö¥ 'local' Î≥ÄÏàòÎ•º Ï∞∏Ï°∞

    inner_function()

outer_function()  # Ï∂úÎ†•: local
```

# ÌååÏù¥Ïç¨ÏóêÏÑúÏùò ÌÅ¥ÎûòÏä§

**ÌÅ¥ÎûòÏä§(Class)**Îäî ÌååÏù¥Ïç¨ÏóêÏÑú Í∞ùÏ≤¥ ÏßÄÌñ• ÌîÑÎ°úÍ∑∏ÎûòÎ∞ç(OOP)ÏùÑ Íµ¨ÌòÑÌïòÎäî Í∏∞Î≥∏Ï†ÅÏù∏ Îã®ÏúÑÏûÖÎãàÎã§. ÌÅ¥ÎûòÏä§Îäî Í∞ùÏ≤¥Î•º Ï†ïÏùòÌïòÎäî **ÌÖúÌîåÎ¶ø(Ï≤≠ÏÇ¨ÏßÑ)**ÏúºÎ°ú, **ÏÜçÏÑ±(Attributes)**Í≥º **Î©îÏÑúÎìú(Methods)**Î•º Ìè¨Ìï®Ìï† Ïàò ÏûàÏäµÎãàÎã§. ÌÅ¥ÎûòÏä§Î•º ÏÇ¨Ïö©ÌïòÎ©¥ ÏΩîÎìúÏùò Ïû¨ÏÇ¨Ïö©ÏÑ±ÏùÑ ÎÜíÏù¥Í≥†, Îç∞Ïù¥ÌÑ∞Î•º Îçî Íµ¨Ï°∞Ï†ÅÏúºÎ°ú Í¥ÄÎ¶¨Ìï† Ïàò ÏûàÏäµÎãàÎã§.

1. ÌÅ¥ÎûòÏä§ Ï†ïÏùòÏôÄ Í∞ùÏ≤¥ ÏÉùÏÑ±
* ÌÅ¥ÎûòÏä§Îäî class ÌÇ§ÏõåÎìúÎ•º ÏÇ¨Ïö©ÌïòÏó¨ Ï†ïÏùòÎê©ÎãàÎã§.
* ÌÅ¥ÎûòÏä§Î•º Ï†ïÏùòÌïú ÌõÑÏóêÎäî Ïù¥Î•º Í∏∞Î∞òÏúºÎ°ú Í∞ùÏ≤¥Î•º ÏÉùÏÑ±Ìï† Ïàò ÏûàÏäµÎãàÎã§. Í∞ùÏ≤¥Îäî ÌÅ¥ÎûòÏä§Ïùò Ïù∏Ïä§ÌÑ¥Ïä§(instance)ÏûÖÎãàÎã§.

ÌÅ¥ÎûòÏä§ Ï†ïÏùò ÏòàÏãú:

```python
class Person:
    # ÌÅ¥ÎûòÏä§Ïùò ÏÉùÏÑ±Ïûê (Í∞ùÏ≤¥ ÏÉùÏÑ± Ïãú Ìò∏Ï∂ú)
    def __init__(self, name, age):
        self.name = name  # Ïù∏Ïä§ÌÑ¥Ïä§ Î≥ÄÏàò
        self.age = age    # Ïù∏Ïä§ÌÑ¥Ïä§ Î≥ÄÏàò

    # Î©îÏÑúÎìú Ï†ïÏùò
    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")
```

Í∞ùÏ≤¥ ÏÉùÏÑ±:

* __init__ Î©îÏÑúÎìúÎäî **ÏÉùÏÑ±Ïûê(Constructor)**Î°ú, Í∞ùÏ≤¥Í∞Ä ÏÉùÏÑ±Îê† Îïå Ìò∏Ï∂úÎêòÎ©∞, Í∞ùÏ≤¥Ïùò Ï¥àÍ∏∞ ÏÉÅÌÉúÎ•º ÏÑ§Ï†ïÌï©ÎãàÎã§.
* selfÎäî ÌÅ¥ÎûòÏä§Ïùò Ïù∏Ïä§ÌÑ¥Ïä§Î•º Í∞ÄÎ¶¨ÌÇ§Î©∞, Ïù∏Ïä§ÌÑ¥Ïä§ Î≥ÄÏàòÎ•º Ï†ïÏùòÌï† Îïå ÏÇ¨Ïö©Îê©ÎãàÎã§. selfÎäî ÌååÏù¥Ïç¨ÏóêÏÑú ÏûêÎèôÏúºÎ°ú Ï†ÑÎã¨ÎêòÏßÄÎßå, Î©îÏÑúÎìúÎ•º Ï†ïÏùòÌï† ÎïåÎäî Î™ÖÏãúÏ†ÅÏúºÎ°ú Ï≤´ Î≤àÏß∏ Ïù∏ÏàòÎ°ú ÏÑ†Ïñ∏Ìï¥Ïïº Ìï©ÎãàÎã§.

```python
# Person ÌÅ¥ÎûòÏä§Ïùò Í∞ùÏ≤¥ ÏÉùÏÑ±
person1 = Person("Alice", 30)
person2 = Person("Bob", 25)

# Í∞ùÏ≤¥Ïùò Î©îÏÑúÎìú Ìò∏Ï∂ú
person1.greet()  # Ï∂úÎ†•: Hello, my name is Alice and I am 30 years old.
person2.greet()  # Ï∂úÎ†•: Hello, my name is Bob and I am 25 years old.
```

2. ÌÅ¥ÎûòÏä§ ÏÜçÏÑ±Í≥º Ïù∏Ïä§ÌÑ¥Ïä§ ÏÜçÏÑ±
* ÌÅ¥ÎûòÏä§ ÏÜçÏÑ±(Class Attributes): Î™®Îì† Ïù∏Ïä§ÌÑ¥Ïä§ÏóêÏÑú Í≥µÏú†ÎêòÎäî Î≥ÄÏàò. ÌÅ¥ÎûòÏä§ Ï†ÑÏ≤¥ÏóêÏÑú ÎèôÏùºÌïú Í∞íÏùÑ Ïú†ÏßÄÌï©ÎãàÎã§.
* Ïù∏Ïä§ÌÑ¥Ïä§ ÏÜçÏÑ±(Instance Attributes): Í∞Å Ïù∏Ïä§ÌÑ¥Ïä§Î≥ÑÎ°ú Í≥†Ïú†Ìïú Í∞íÏùÑ Í∞ÄÏßÄÎäî Î≥ÄÏàò. Í∞ùÏ≤¥ÎßàÎã§ ÎèÖÎ¶ΩÏ†ÅÏù∏ Í∞íÏùÑ Í∞ÄÏßëÎãàÎã§.

```python
class Dog:
    # ÌÅ¥ÎûòÏä§ ÏÜçÏÑ± (Î™®Îì† Ïù∏Ïä§ÌÑ¥Ïä§Í∞Ä Í≥µÏú†)
    species = "Canis lupus familiaris"

    def __init__(self, name, age):
        # Ïù∏Ïä§ÌÑ¥Ïä§ ÏÜçÏÑ± (Í∞Å Ïù∏Ïä§ÌÑ¥Ïä§Í∞Ä Í≥†Ïú†ÌïòÍ≤å Í∞ÄÏßê)
        self.name = name
        self.age = age

# Í∞ùÏ≤¥ ÏÉùÏÑ±
dog1 = Dog("Buddy", 5)
dog2 = Dog("Charlie", 3)

# ÌÅ¥ÎûòÏä§ ÏÜçÏÑ± Ï†ëÍ∑º
print(dog1.species)  # Ï∂úÎ†•: Canis lupus familiaris
print(dog2.species)  # Ï∂úÎ†•: Canis lupus familiaris

# Ïù∏Ïä§ÌÑ¥Ïä§ ÏÜçÏÑ± Ï†ëÍ∑º
print(dog1.name)  # Ï∂úÎ†•: Buddy
print(dog2.name)  # Ï∂úÎ†•: Charlie
```

3. Î©îÏÑúÎìú

* Ïù∏Ïä§ÌÑ¥Ïä§ Î©îÏÑúÎìú: Í∞ùÏ≤¥Ïùò ÏÉÅÌÉú(Ïù∏Ïä§ÌÑ¥Ïä§ Î≥ÄÏàò)Î•º Î≥ÄÍ≤ΩÌïòÍ±∞ÎÇò ÏùΩÎäî Î©îÏÑúÎìúÏûÖÎãàÎã§. selfÎ•º Ï≤´ Î≤àÏß∏ Ïù∏ÏàòÎ°ú Î∞õÏïÑÏïº Ìï©ÎãàÎã§.
* ÌÅ¥ÎûòÏä§ Î©îÏÑúÎìú: ÌÅ¥ÎûòÏä§ ÏûêÏ≤¥Ïóê Ï†ÅÏö©ÎêòÎäî Î©îÏÑúÎìúÎ°ú, @classmethod Îç∞ÏΩîÎ†àÏù¥ÌÑ∞Î•º ÏÇ¨Ïö©ÌïòÍ≥† Ï≤´ Î≤àÏß∏ Ïù∏ÏàòÎ°ú cls(ÌÅ¥ÎûòÏä§)Î•º Î∞õÏäµÎãàÎã§.
* Ï†ïÏ†Å Î©îÏÑúÎìú: ÌÅ¥ÎûòÏä§ÎÇò Ïù∏Ïä§ÌÑ¥Ïä§ÏôÄ Í¥ÄÎ†® ÏóÜÏù¥ ÎèôÏûëÌïòÎäî Î©îÏÑúÎìúÏûÖÎãàÎã§. @staticmethod Îç∞ÏΩîÎ†àÏù¥ÌÑ∞Î•º ÏÇ¨Ïö©ÌïòÏó¨ Ï†ïÏùòÌïòÎ©∞, selfÎÇò clsÎ•º Î∞õÏßÄ ÏïäÏäµÎãàÎã§.

```python
class Circle:
    pi = 3.14159  # ÌÅ¥ÎûòÏä§ ÏÜçÏÑ±

    def __init__(self, radius):
        self.radius = radius  # Ïù∏Ïä§ÌÑ¥Ïä§ ÏÜçÏÑ±

    # Ïù∏Ïä§ÌÑ¥Ïä§ Î©îÏÑúÎìú
    # Ïù∏Ïä§ÌÑ¥Ïä§ Î©îÏÑúÎìúÎäî Ïù∏Ïä§ÌÑ¥Ïä§ Î≥ÄÏàòÎ•º Ïù¥Ïö©Ìï¥ ÎèôÏûëÌïòÎ©∞, Í∞ùÏ≤¥ÎßàÎã§ Îã§Î•∏ ÎèôÏûëÏùÑ ÏàòÌñâÌï† Ïàò ÏûàÏäµÎãàÎã§.
    def area(self):
        return Circle.pi * self.radius ** 2

    # ÌÅ¥ÎûòÏä§ Î©îÏÑúÎìú
    # ÌÅ¥ÎûòÏä§ Î©îÏÑúÎìúÎäî ÌÅ¥ÎûòÏä§ ÏûêÏ≤¥Î•º Ï≤´ Î≤àÏß∏ Ïù∏ÏûêÎ°ú Î∞õÎäî Î©îÏÑúÎìúÏûÖÎãàÎã§. ÌÅ¥ÎûòÏä§ Î©îÏÑúÎìúÎ•º Ï†ïÏùòÌï† ÎïåÎäî Î©îÏÑúÎìú ÏúÑÏóê @classmethod Îç∞ÏΩîÎ†àÏù¥ÌÑ∞Î•º ÏÇ¨Ïö©Ìï©ÎãàÎã§.
    # ÌÅ¥ÎûòÏä§ Î©îÏÑúÎìúÏùò Ï≤´ Î≤àÏß∏ Ïù∏ÏûêÎäî **cls**Î°ú, Ïù¥Îäî Ìï¥Îãπ Î©îÏÑúÎìúÎ•º Ìò∏Ï∂úÌïú ÌÅ¥ÎûòÏä§ Í∞ùÏ≤¥Î•º ÎÇòÌÉÄÎÉÖÎãàÎã§.
    # ÌÅ¥ÎûòÏä§ Î©îÏÑúÎìúÎäî Ï£ºÎ°ú ÌÅ¥ÎûòÏä§ ÏÉÅÌÉúÎÇò ÌÅ¥ÎûòÏä§ Î≥ÄÏàòÎ•º ÏàòÏ†ïÌïòÍ±∞ÎÇò Ï†ëÍ∑ºÌï† Îïå ÏÇ¨Ïö©Ìï©ÎãàÎã§
    @classmethod
    def from_diameter(cls, diameter):
        return cls(diameter / 2)

    # Ï†ïÏ†Å Î©îÏÑúÎìú
    # Ï†ïÏ†Å Î©îÏÑúÎìúÎäî ÌÅ¥ÎûòÏä§ÏôÄ Î¨¥Í¥ÄÌïú ÎèÖÎ¶ΩÏ†ÅÏù∏ ÏûëÏóÖÏùÑ ÏàòÌñâÌï©ÎãàÎã§.
    @staticmethod
    def is_positive(number):
        return number > 0

# Ïù∏Ïä§ÌÑ¥Ïä§ ÏÉùÏÑ±
circle = Circle(5)
print(circle.area())  # Ï∂úÎ†•: 78.53975

# ÌÅ¥ÎûòÏä§ Î©îÏÑúÎìúÎ°ú Ïù∏Ïä§ÌÑ¥Ïä§ ÏÉùÏÑ±
# ÌÅ¥ÎûòÏä§ Î©îÏÑúÎìúÎäî ÌÅ¥ÎûòÏä§ Ï†ÑÏ≤¥Ïóê ÏòÅÌñ•ÏùÑ ÎØ∏ÏπòÎ©∞, Î≥¥ÌÜµ Í∞ùÏ≤¥ ÏÉùÏÑ±Í≥º Í¥ÄÎ†®Îêú Î©îÏÑúÎìúÏóêÏÑú ÎßéÏù¥ ÏÇ¨Ïö©Îê©ÎãàÎã§.
circle_from_diameter = Circle.from_diameter(10)
print(circle_from_diameter.radius)  # Ï∂úÎ†•: 5.0

# Ï†ïÏ†Å Î©îÏÑúÎìú Ìò∏Ï∂ú
# Ï†ïÏ†Å Î©îÏÑúÎìúÎäî ÌÅ¥ÎûòÏä§ÏôÄ Î¨¥Í¥ÄÌïú ÎèÖÎ¶ΩÏ†ÅÏù∏ ÏûëÏóÖÏùÑ ÏàòÌñâÌï©ÎãàÎã§.
print(Circle.is_positive(-5))  # Ï∂úÎ†•: False
```

```python
class MyClass:
    class_variable = 0

    @classmethod
    def class_method(cls):
        cls.class_variable += 1
        print(f'ÌÅ¥ÎûòÏä§ Î≥ÄÏàò: {cls.class_variable}')

MyClass.class_method()  # ÌÅ¥ÎûòÏä§ Î©îÏÑúÎìú Ìò∏Ï∂ú
MyClass.class_method()  # ÌÅ¥ÎûòÏä§ Î©îÏÑúÎìú Ìò∏Ï∂ú
myclass = MyClass()
print(myclass.class_variable)
```

4. ÏÉÅÏÜç(Inheritance)
* ÏÉÅÏÜçÏùÄ Í∏∞Ï°¥ ÌÅ¥ÎûòÏä§Î•º Ïû¨ÏÇ¨Ïö©ÌïòÏó¨ ÏÉàÎ°úÏö¥ ÌÅ¥ÎûòÏä§Î•º ÎßåÎìúÎäî Î∞©Î≤ïÏûÖÎãàÎã§. ÏÉÅÏÜçÎ∞õÏùÄ ÌÅ¥ÎûòÏä§Îäî Î∂ÄÎ™® ÌÅ¥ÎûòÏä§Ïùò Î™®Îì† ÏÜçÏÑ±Í≥º Î©îÏÑúÎìúÎ•º Î¨ºÎ†§Î∞õÏúºÎ©∞, Ïù¥Î•º ÌôïÏû•ÌïòÍ±∞ÎÇò Ïû¨Ï†ïÏùòÌï† Ïàò ÏûàÏäµÎãàÎã§.
* Î∂ÄÎ™® ÌÅ¥ÎûòÏä§Î•º ÏÉÅÏÜçÎ∞õÏùÄ ÌÅ¥ÎûòÏä§Î•º ÏûêÏãù ÌÅ¥ÎûòÏä§ÎùºÍ≥† Ìï©ÎãàÎã§.

```python
# Î∂ÄÎ™® ÌÅ¥ÎûòÏä§
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return "Some sound"

# ÏûêÏãù ÌÅ¥ÎûòÏä§ (AnimalÏùÑ ÏÉÅÏÜç)
class Dog(Animal):
    def speak(self):
        return "Woof!"

# ÏûêÏãù ÌÅ¥ÎûòÏä§ (AnimalÏùÑ ÏÉÅÏÜç)
class Cat(Animal):
    def speak(self):
        return "Meow!"

# Animal Í∞ùÏ≤¥ ÏÉùÏÑ±
ani = Animal("dog")
print(ani.name)
print(ani.speak())

# Í∞ùÏ≤¥ ÏÉùÏÑ±
dog = Dog("Buddy")
cat = Cat("Whiskers")

print(dog.name)  # Ï∂úÎ†•: Buddy
print(dog.speak())  # Ï∂úÎ†•: Woof!
print(cat.speak())  # Ï∂úÎ†•: Meow!
```

* DogÏôÄ Cat ÌÅ¥ÎûòÏä§Îäî Animal ÌÅ¥ÎûòÏä§Î•º ÏÉÅÏÜçÎ∞õÏïÑ Î∂ÄÎ™® ÌÅ¥ÎûòÏä§Ïùò ÏÜçÏÑ±ÏùÑ Î¨ºÎ†§Î∞õÏäµÎãàÎã§.
* speak() Î©îÏÑúÎìúÎäî Í∞Å ÏûêÏãù ÌÅ¥ÎûòÏä§ÏóêÏÑú Ïû¨Ï†ïÏùò(override) ÎêòÏñ¥ Í≥†Ïú†Ìïú ÎèôÏûëÏùÑ ÏàòÌñâÌï©ÎãàÎã§.

5. Îã§ÌòïÏÑ±(Polymorphism)
* Îã§ÌòïÏÑ±ÏùÄ ÎèôÏùºÌïú Ïù¥Î¶ÑÏùò Î©îÏÑúÎìúÍ∞Ä Îã§Î•∏ ÌÅ¥ÎûòÏä§ÏóêÏÑú Îã§Î•∏ Î∞©ÏãùÏúºÎ°ú ÎèôÏûëÌïòÎäî Í≤ÉÏùÑ ÏùòÎØ∏Ìï©ÎãàÎã§.
* ÏòàÎ•º Îì§Ïñ¥, Î∂ÄÎ™® ÌÅ¥ÎûòÏä§ÏóêÏÑú Ï†ïÏùòÎêú Î©îÏÑúÎìúÎ•º ÏûêÏãù ÌÅ¥ÎûòÏä§ÏóêÏÑú Ïû¨Ï†ïÏùò(Ïò§Î≤ÑÎùºÏù¥Îî©)Ìï† Ïàò ÏûàÏúºÎ©∞, ÏÑúÎ°ú Îã§Î•∏ ÌÅ¥ÎûòÏä§Ïùò Í∞ùÏ≤¥Îì§Ïù¥ ÎèôÏùºÌïú Ïù∏ÌÑ∞ÌéòÏù¥Ïä§Î°ú ÏÉÅÌò∏ÏûëÏö©Ìï† Ïàò ÏûàÏäµÎãàÎã§.

```python
def animal_sound(animal):
    print(animal.speak())

dog = Dog("Buddy")
cat = Cat("Whiskers")

animal_sound(dog)  # Ï∂úÎ†•: Woof!
animal_sound(cat)  # Ï∂úÎ†•: Meow!

```

* animal_sound() Ìï®ÏàòÎäî DogÏôÄ Cat Í∞ùÏ≤¥Î•º ÎèôÏùºÌïòÍ≤å Ï≤òÎ¶¨ÌïòÏßÄÎßå, Í∞ÅÍ∞ÅÏùò Í∞ùÏ≤¥ÏóêÏÑú Ïû¨Ï†ïÏùòÎêú speak() Î©îÏÑúÎìúÎ•º Ìò∏Ï∂úÌï©ÎãàÎã§.

6. Ï∫°ÏäêÌôî(Encapsulation)Í≥º Ï†ëÍ∑º Ï†úÏñ¥
* Ï∫°ÏäêÌôîÎäî Í∞ùÏ≤¥Ïùò ÎÇ¥Î∂Ä Îç∞Ïù¥ÌÑ∞Î•º Ïô∏Î∂ÄÏóêÏÑú ÏßÅÏ†ë Ï†ëÍ∑ºÌïòÏßÄ Î™ªÌïòÎèÑÎ°ù Ï†úÌïúÌïòÎäî Í∏∞Î≤ïÏûÖÎãàÎã§.
* ÌååÏù¥Ïç¨ÏóêÏÑú Î≥ÄÏàòÎÇò Î©îÏÑúÎìú Ïù¥Î¶Ñ ÏïûÏóê Î∞ëÏ§Ñ _ ÎòêÎäî **Ïù¥Ï§ë Î∞ëÏ§Ñ __**ÏùÑ Î∂ôÏó¨ÏÑú Ï†ëÍ∑º Ï†úÏñ¥Î•º Ìï† Ïàò ÏûàÏäµÎãàÎã§.
* Îã®Ïùº Î∞ëÏ§Ñ _: ÎÇ¥Î∂Ä ÏÇ¨Ïö©ÏùÑ ÏùòÎØ∏ÌïòÎäî Í¥ÄÎ°ÄÎ°ú, ÏßÅÏ†ë ÏÇ¨Ïö©ÌïòÏßÄ ÎßêÎùºÎäî ÏùòÎØ∏.
* Ïù¥Ï§ë Î∞ëÏ§Ñ __: **Ïù¥Î¶Ñ ÎßπÍ∏ÄÎßÅ(name mangling)**ÏùÑ ÌÜµÌï¥ Ïô∏Î∂ÄÏóêÏÑú Ï†ëÍ∑ºÌïòÏßÄ Î™ªÌïòÍ≤å Ìï©ÎãàÎã§.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self._age = age  # Í¥ÄÎ°ÄÏÉÅ ÎπÑÍ≥µÍ∞ú Î≥ÄÏàò

    def get_age(self):
        return self._age

    def set_age(self, age):
        if age > 0:
            self._age = age

# Í∞ùÏ≤¥ ÏÉùÏÑ±
person = Person("John", 30)

# ÎπÑÍ≥µÍ∞ú Î≥ÄÏàò Ï†ëÍ∑º (Í¥ÄÎ°ÄÏÉÅ ÏßÅÏ†ë Ï†ëÍ∑ºÌïòÏßÄ ÏïäÏùå)
print(person._age)  # Ï∂úÎ†•: 30

# GetterÏôÄ Setter ÏÇ¨Ïö©
person.set_age(35)
print(person.get_age())  # Ï∂úÎ†•: 35
```
